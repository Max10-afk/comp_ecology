```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Graphs

- $\frac{dT}{dt} = T + h \cdot (T_{max} - T) - (T - T_{min}) \cdot V^2 \cdot W$
- $\frac{dW}{dt} = W$
- $\frac{dV}{dt} = V + p \cdot (1 - V) - V \cdot \frac{1}{1 + \exp(-(T_{min} / T_{max})\cdot (T - T_{crit}))}$

```python
t_max = 50
t_min= 20
t_crit = 40
h = 0.1
W = 0.5
p = 0.1
```

## x: Vegetation, y: Temperature

```python
T = 35
x = np.linspace(0,1,100)
y = T + h * (t_max - T) - (T - t_min) * W * x**2

plt.plot(x,y,"darkorange",label="T")
plt.ylabel("Temperature")
plt.xlabel("Vegetation coverage")
plt.hlines(t_max,0,1,"r","--",label="$T_{max}$")
plt.hlines(t_min,0,1,"b","--",label="$T_{min}$")
plt.legend(loc="best")
```

## x: Temperature y: Plants

```python
V = 0.6
x = np.linspace(20,50,100)
y = V + p * (1-V) - V * (1+np.exp(-(t_min / t_max)*(x - t_crit)))**-1

plt.plot(x,y,"g")
plt.xlabel("Temperature")
plt.ylabel("Vegetation coverage")
```

## x: Water bodies y: Temperature

```python
V = 0.5
x = np.linspace(0,1,100)
y = T + h * (t_max - T) - (T - t_min) * x * V**2

plt.plot(x,y)
plt.ylabel("Temperature")
plt.xlabel("Water bodies")
```

## x: planting y: vegetation coverage

```python
V = 0.5
T = 35
x = np.linspace(0,1,100)
y = V + x * (1-V) - V * (1+np.exp(-(t_min / t_max)*(T - t_crit)))**-1
plt.plot(x,y,"g")
plt.xlabel("Plannted percentage")
plt.ylabel("Vegetation coverage")
```
